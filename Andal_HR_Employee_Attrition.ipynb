**For Upper and Lower Fence**

```python
q1 = df1['Daily_Salary'].describe()['25%']

# Get third quartile
q3 = df1['Daily_Salary'].describe()['75%']

# Get interquartile range (IQR)
iqr = q3 - q1

# Get lower fence
lf = q1 - (1.5 * iqr)

# Get upper fence
uf = q3 + (1.5 * iqr)

print("Upper fence: ", uf)
print("Lower fence: ", lf)
```

```python
df1[(df1['Daily_Salary'] > lf) & (df1['Daily_Salary'] < uf)]
```
