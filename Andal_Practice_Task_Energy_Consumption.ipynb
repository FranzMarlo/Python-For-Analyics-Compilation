# **Importing of Libraries**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from time import time
```

# **Data Loading**

```python
andal = pd.read_csv('/content/Energy Consumption Data.csv')
andal
```

# **Data Cleaning**

# Converting Data Types

```python
andal['Date_Time'] = pd.to_datetime(andal['Date'] + ' '+ andal['Time'])
andal['AD'] = pd.to_numeric(andal['AD'], errors ='coerce')
andal['Ave_Temp'] = pd.to_numeric(andal['AVERAGE TEMPERATURE'], errors = 'coerce')
andal['Ave_Rainfall'] = pd.to_numeric(andal['AVERAGE RAINFALL'], errors = 'coerce')
andal
```

# Removing Null Values

```python
andal = andal.dropna()
andal
```

# Extract Year, Quarter, Month, and Day

```python
andal = andal.copy()
andal['Year'] = andal['Date_Time'].apply(lambda x: x.year)
andal['Quarter'] = andal['Date_Time'].apply(lambda x: x.quarter)
andal['Month'] = andal['Date_Time'].apply(lambda x: x.month)
andal['Day'] = andal['Date_Time'].apply(lambda x: x.day)
andal
```

# Re Arranging Table


```python
andal = andal.loc[:, ['Date_Time', 'Year', 'Quarter', 'Month', 'Day', 'Ave_Temp', 'Ave_Rainfall', 'AD']]
andal.sort_values('Date_Time', inplace = True, ascending = True)
andal = andal.reset_index(drop = True)
andal
```

```python
andal.describe()
```

```python
andal['Weekday'] = andal.apply(lambda row: row['Date_Time'].weekday(), axis = 1)
andal['Weekday'] = (andal['Weekday'] <5). astype(int)
```

```python
print(andal.shape)
print(andal.Date_Time.min())
print(andal.Date_Time.max())
andal
```

```python
from scipy import stats
stat, p = stats.normaltest(andal.AD)

print ('Statistics = %.3f , p = %.3f' %(stat, p))
alpha = 0.05

if p > alpha:
  print('Data looks Gaussian')
else:
  print('Data does not Look Gaussian')
```

# **Kurtosis and Skewness**

```python
plt.figure(figsize=(10, 6))
plt.gcf().set_facecolor('#2E2E2E')
ax = sns.histplot(andal.AD, color='#3DC2EC', edgecolor='white')
ax.set_facecolor('#1C1C1C')
ax.set_title('Distribution', color='white')
ax.set_xlabel('Value', color='white')
ax.set_ylabel('Frequency', color='white')
ax.tick_params(axis='x', colors='white')
ax.tick_params(axis='y', colors='white')

plt.show()
print('Kurtosis: {}' .format (stats.kurtosis(andal.AD)))
print('Skewness: {}' .format (stats.skew(andal.AD)))
```

Interpretation:

Skewness: Data is *positively skewed* (right skew).

Kurtosis: Data has *light tails* (platykurtic).

# **Visualization**

**Energy Consumption By Year**

```python
import matplotlib.ticker as tkr

fig = plt.figure(figsize=(12, 20))
fig.patch.set_facecolor('#1C1C1C')
fig.subplots_adjust(hspace=.4)

fig1 = fig.add_subplot(5, 1, 1)
fig1.set_facecolor('#1C1C1C')
fig1.bar(andal1['AD'].resample('A').mean().index, andal1['AD'].resample('A').mean(), width=150, color='#3DC2EC', edgecolor='white')

fig1.set_title('Yearly', color='white')
fig1.tick_params(axis='x', colors='white')
fig1.tick_params(axis='y', colors='white')

plt.show()

```

**Analysis:**

The trend analysis from 2013 to 2018 shows a general increase in energy consumption, which peaks in 2018 due to a combination of factors. This has increased significantly from 2013 to 2018, surging from nearly 6,000 energy consumption to around 7,000 energy consumption by 2018. The highest for this period, which was in this year, 2018, is then depicted through a bar graph to show the gigantic difference between years. In this data, the average temperature and rainfall of the period are also gauged to determine the energy consumption per year. It seems like starting in the year 2013, the energy consumption begins to add a minimal percent; this rapidly grows until it reaches the year 2018, wherein the energy consumption increases by about 16%. As average temperature was indicated and as evidence, it has been greatly affected by a shift in climate change in global temperatures. Between the years 2013 and 2018, the world temperatures kept rising. In 2018, it was the fourth hottest, with an average temperature of 0.79 °C above average during the 20th century. This heightened demand for energy used by cooling systems in homes and commercial buildings. It eventually drove a sharp rise in electricity demand. The same trend is seen across the globe. Therefore, this could be the reason for the drastic increase in energy consumption per year, and as we assume, this could contagiously increase as time goes by.

**Energy Consumption Through The Years Quarterly**

```python
plt.style.use('dark_background')
fig = plt.figure(figsize=(12, 20))
fig.subplots_adjust(hspace=0.4)
fig2 = fig.add_subplot(5, 1, 2)
fig2.plot(andal1['AD'].resample('Q').mean(), color='#3dc2ec', linewidth=5)
fig2.set_title('Quarterly', color='white')
fig2.tick_params(axis='both', which='major', colors='white')
plt.ylabel('AD', color='white')
plt.xlabel('Date', color='white')
plt.tight_layout()
plt.show()
```

**Analysis:**

In general, energy consumption is on obvious uptrend from 2012 to 2018 and peaks in the years around 2017 and 2018. There are apparent fluctuations throughout each year and most likely reflect changes in seasons. The highest consumption point is around late 2017 or early 2018, reaching near 8,000 energy consumption. The lowest consumption is around 2012 and early 2013, staying below 6,000 energy consumption. It is growing every year, with a remarkable growth from 2016 to 2017. Factors responsible for these trends include seasonality, economic growth, technological growth, and efficiency in energy use. The graph concludes that energy consumption has gone through stable growth from 2012 to 2018 with clear quarterly fluctuations, which have likely been caused by a combination of economic growth, urbanization, climate condition, and technological growth in these years.

**Energy Consumption Through The Years Weekly**

**Analysis:**

The line graph reveals that the trend for the consumption of energy on a weekly basis was cyclical in nature, peaking in the middle to late parts of the years and dipping in the early months. There is a seasonal peaking, especially from 2015 onwards, which shows big spikes in consumption during certain weeks. Notwithstanding, there is an upward trend of energy consumption across time, higher in later years than in earlier years, probably due to economic or population increase. This also reflects a general upward tendency with a steady rise over time within these weekly peaks in economic growth, urbanization, and a change in energy policies. In 2018, energy use has surged upwards, and consecutive weeks reached a higher peak compared to 2012. It simply underlines the rising reliance of modern life and economic activities on electricity.
